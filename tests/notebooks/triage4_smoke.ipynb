# triage4 smoke test

Validate package import and a deterministic scheduling run.

```python
from triage4 import TRIAGE4Config, TRIAGE4Scheduler

cfg = TRIAGE4Config(high_zone_max=1, standard_zone_max=3, service_rate=20.0)
scheduler = TRIAGE4Scheduler(cfg, scheduler_seed=42)
result = scheduler.schedule(
    arrival_times=[0.0, 0.01, 0.02, 0.03],
    device_ids=['d1', 'd2', 'd1', 'd3'],
    zone_priorities=[0, 2, 5, 1],
    is_alarm=[False, True, False, False],
)

assert len(result.waiting_times) == 4
assert len(result.e2e_times) == 4
assert len(result.priorities) == 4
print('smoke-ok', list(result.priorities))
```
